```python
from IPython.display import clear_output
import time

def set_time():
    while True:
        try:
            hours = int(input("Enter the hour (0-12): "))
            if hours < 0 or hours > 12:
                raise ValueError
            break
        except ValueError:
            print("Invalid input. Please enter a number between 0 and 12.")

    while True:
        try:
            minutes = int(input("Enter the minutes (0-59): "))
            if minutes < 0 or minutes > 59:
                raise ValueError
            break
        except ValueError:
            print("Invalid input. Please enter a number between 0 and 59.")

    while True:
        try:
            seconds = int(input("Enter the seconds (0-59): "))
            if seconds < 0 or seconds > 59:
                raise ValueError
            break
        except ValueError:
            print("Invalid input. Please enter a number between 0 and 59.")

    return hours, minutes, seconds

def digital_clock():
    hours, minutes, seconds = set_time()

    while True:
        # Clear output
        clear_output(wait=True)

        # Print the current time
        time_str = time.strftime("%I:%M:%S %p", time.localtime())
        print(f"Digital Clock: {time_str}")

        # Update the time
        seconds += 1
        if seconds == 60:
            seconds = 0
            minutes += 1
        if minutes == 60:
            minutes = 0
            hours += 1
        if hours == 13:
            hours = 1

        # Wait for one second
        time.sleep(1)

digital_clock()
```
