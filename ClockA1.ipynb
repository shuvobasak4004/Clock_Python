```python
import datetime
import IPython.display as display
import pytz
import time

def draw_clock(h, m, s):
    clock = [
        "      XII",
        "   I        XI",
        "II             X",
        "  III         IX",
        "     IV    VIII",
        "        V VI",
        "       VII"
    ]

    # Calculate the angles for hour, minute, and second hands
    hour_angle = (h % 12 + m / 60 + s / 3600) * 30
    minute_angle = (m + s / 60) * 6
    second_angle = s * 6

    # Draw the hour hand
    hour_hand = f"\033[1;32mH ({h:02})\033[0m"
    clock[3] = clock[3][:7] + hour_hand.center(5) + clock[3][12:]

    # Draw the minute hand
    minute_hand = f"\033[1;34mM ({m:02})\033[0m"
    clock[5] = clock[5][:7] + minute_hand.center(5) + clock[5][12:]

    # Draw the second hand
    second_hand = f"\033[1;31mS ({s:02})\033[0m"
    clock[2] = clock[2][:7] + second_hand.center(5) + clock[2][12:]

    # Print the clock
    display.clear_output(wait=True)
    print("\n".join(clock))

# Set the time zone to Bangladesh
bd_tz = pytz.timezone('Asia/Dhaka')

while True:
    # Get current time in Bangladesh time zone
    current_time = datetime.datetime.now(tz=bd_tz)
    h, m, s = current_time.hour, current_time.minute, current_time.second

    # Draw the clock
    draw_clock(h, m, s)

    # Wait for one second
    time.sleep(1)
```
