```python
import datetime
from IPython.display import clear_output
import time

while True:
    # Clear output
    clear_output(wait=True)

    # Get current time
    current_time = datetime.datetime.now().strftime("%I:%M:%S %p")

    # Print the current time
    print(f"Digital Clock: {current_time}")

    # Wait for one second
    time.sleep(1)
```
